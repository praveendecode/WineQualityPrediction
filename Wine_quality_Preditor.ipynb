````python
# Dropping Irrelevant Columns

df1 = df[['fixed acidity',
          'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides',
        'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates',
        'alcohol', 'quality',
        # 'Id'
       
       
       
       ]].copy()
````

````python
# Adding Quality label

df1["quality_label"] = df1["quality"].apply(lambda x :1 if x>6 else 0)

df1["label"] = df1["quality"].apply(lambda x:"Good"  if x > 6 else "Bad")

#note : Here good quality wines are mentioned by 1 with threshold 6 , bad quality is mentioned by 0.
````

````python
df1.columns =['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
       'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
       'pH_level', 'sulphates', 'alcohol', 'quality', 'quality_label', "label"]
````

````python
df2 = df1[['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
       'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
       'pH_level', 'sulphates', 'alcohol', 
       
       #'quality',
        'quality_label',
       #'label'
       ]]

# for regressor model is used 
````

## **Regression -- Linear Regressor Model**

````python
# Dividing Data into Input(x) & Output(y) variables 

x = df2.iloc[:,:]     # input

y = df1["quality"].values       # output


# Dimensions of x , y

print(f"The dimension of x :{x.ndim} \nThe dimension of y :{y.ndim}")
````

````python
# Train and Test splits :
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
````

````python
# Scaling the input data 

scaler = MinMaxScaler() 


Scaled_x_train = scaler.fit_transform(x_train)


Scaled_x_test = scaler.transform(x_test)
````

````python
# Linear Regression Algorithm

model = LinearRegression()

# model fit

model.fit(Scaled_x_train,y_train)
````

````python
# Output Prediction

linre_prediction = model.predict(Scaled_x_test)
````

````python
# Actual Output

print(y_test)

print()
# Predicted output 

print(linre_prediction)
````

````python
# Individual Prediction 
model.predict([[0.35454545, 0.10958904 ,0.56 ,      0.06164384, 0.13355593, 0.50746269,
 0.35661765 ,0.52018279 ,0.44094488 ,0.25149701, 0.375 ,     0.8      ] ])
````

````python
# Accuracy of the model 


mse = mean_squared_error(y_test, linre_prediction)
r2 = r2_score(y_test, linre_prediction)



print(f'Mean Squared Error:{mse:.3f}')

print(f"R-Squared Score:{r2:.3f}")
````

## **Classification -- Logistic Regressor Model:** 

````python
# Dividing Data into Input(x) & Output(y) variables 

x = df1.iloc[:,:-2].values      # input

y = df1["label"].values       # output


#Dimensions of x , y

print(f"The dimension of x :{x.ndim} \nThe dimension of y :{y.ndim}")
````

````python
# Train and Test splits :
x_train , x_test ,y_train , y_test = train_test_split(x,y,test_size =0.2 ,random_state=42)
````

````python
# Scaling the input data 

scaler = MinMaxScaler() 


Scaled_x_train = scaler.fit_transform(x_train)


Scaled_x_test = scaler.transform(x_test)
````

````python
# Logistic Regression 

model = LogisticRegression()

# model fit

model.fit(Scaled_x_train,y_train)
````

````python
# Output Prediction

logre_prediction = model.predict(Scaled_x_test)
````

````python
# Actual Output

print(y_test)

print()
# Predicted output 

print(logre_prediction)
````

````python
# Accuracy of Our model

print(f"The Accuracy of model is : {accuracy_score(logre_prediction,y_test)*100:.2f}")
````
